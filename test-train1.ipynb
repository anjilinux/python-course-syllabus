```python
class dgx(Exception):
    def __init__(self, age, msg="age is 0 to 120"):
        self.age = age
        self.msg = msg
        super().__init__(self.msg)
    def __str__(self):
        return f'{self.age} -> {self.msg}'
def gpu(age):
    if age < 0 or age > 120:
        raise dgx(age)  
    else: print(f'age set to: {age}')  
try: 
    gpu(150)
except dgx as e:
    print(e)        


```

```python
class cpu(Exception):
    def __init__(self, message):
        self.message = message 
        super().__init__(self.message)
def ram(value):
    if value < 0:
        raise cpu(" value cannot be negative   ")
    else:
        return f"value is {value}"
try:
    re = ram(-5)    
    print(re)
except cpu as e:   
    print(e.message) 

```
